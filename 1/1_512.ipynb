```python
import altair as alt
import pandas as pd
import numpy as np

df = pd.read_csv('sunshine.csv')

conds1 = [
    (df['monthnum']>3) & (df['monthnum']<8),
    (df['monthnum']>9) | (df['monthnum']<2)]
choices1 = ['Summer','Winter']
df['season'] = np.select(conds1, choices1, default='Neither')


conds2 = [
    (df['city']=='Chicago'), (df['city']=='Houston'), (df['city']=='Miami'), 
    (df['city']=='New York'), (df['city']=='San Francisco'),(df['city']=='Seattle')]
choices2 = [10.5,13.3,13.2,8,10,15.2]
df['death_rate'] = np.select(conds2, choices2, default=0)
df['monthly sunshine hours'] = df['sunshine']

trunc_df = df.loc[df['season']!='Neither']

bar_chart = alt.Chart(trunc_df).mark_bar().encode(
    color=alt.Color('season')
        .scale(range=['orange','blue']),
    y=alt.Y('season',axis=None),
    row=alt.Row('city', title=None, header=None),
    x='mean(monthly sunshine hours):Q'
).properties(
    title=''
)

death_chart = alt.Chart(trunc_df).mark_bar(color='red',opacity=0.65, size=25).encode(
    #row=alt.Row('city',header=alt.Header(labelOrient='left'), title=None),
    row=alt.Row('city',header=None, title=None),
    x=alt.X('death_rate:Q', title='Suicide rate (per 100k)')
).properties(
    title="Which cities are best for Seasonal Affective Disorder?"
).properties(
    height=40
)

middle_chart = alt.Chart(trunc_df).mark_text().encode(
    row=alt.Row('city', title=None)
        .header(labelAngle=0, labelAlign='left', labelFontSize=12)
).properties(
    height=40
)
combined_chart = alt.concat(middle_chart, death_chart, bar_chart)
combined_chart = combined_chart.configure_view(strokeWidth=0).configure_axis(grid=False).configure_title(fontSize=15)
#combined_chart = alt.concat(death_chart, bar_chart)
combined_chart.display()
```
